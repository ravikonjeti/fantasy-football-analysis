```python
import csv
import pandas as pd
import numpy as np
```

```python
fantasyTotal = pd.read_csv('seasonFantasy.csv')
dkSalary = pd.read_csv('DKSalaries_LARSFO.csv')
injury = pd.read_csv('InjuryReport.csv')
```

```python
#split names in injury report
injuryName = []
for name in injury['Player']:
    injuryName.append(name.split('\\')[0])
    
injury['Name'] = injuryName
```

```python
#split player name and IDs in fantasyTotal dataframe
playerName = []
playerID = []
for name in fantasyTotal['Player']:
    splitName = str(name).split('\\')
    if len(splitName[0]) > 0:
        playerName.append(splitName[0])
    else:
        playerName.append('NA')
        
    if len(splitName[1]) > 0:
        playerID.append(splitName[1])
    else:
        playerID.append('NA')        
    
fantasyTotal['Name'] = playerName
fantasyTotal['ID'] = playerID
```

```python
#Add injuries to the fantasyTotal dataframe
injList = []
for name in fantasyTotal['Name']:
    status = injury[injury['Name'] == name]['Class']
    if len(status) > 0:
        injList.append(status.iloc[0])
    else:
        injList.append('NA')
    
fantasyTotal['Injury'] = injList
```

```python
fantasyTotal = fantasyTotal[(fantasyTotal['Team'] == 'LAR') | (fantasyTotal['Team'] == 'SFO')]
```

```python
#add salaries to the fantasyTotal dataframe
dkCPT = dkSalary[dkSalary['Roster Position'] == 'CPT']
dkFlex = dkSalary[dkSalary['Roster Position'] == 'FLEX']

salaryListCPT = []
salaryListFlex = []

for player in fantasyTotal['Name']:
    salCPT = dkCPT[dkCPT['Name'] == player]['Salary']
    salFlex = dkFlex[dkFlex['Name'] == player]['Salary']
    if len(salCPT) > 0:
        salaryListCPT.append(salCPT.iloc[0])
    else:
        salaryListCPT.append(0)
    if len(salFlex) > 0:
        salaryListFlex.append(salFlex.iloc[0])
    else:
        salaryListFlex.append(0)
    
fantasyTotal['Salary CPT'] = salaryListCPT
fantasyTotal['Salary Flex'] = salaryListFlex
```

```python
dkCPT = dkSalary[dkSalary['Roster Position'] == 'CPT']
dkFlex = dkSalary[dkSalary['Roster Position'] == 'FLEX']
```

```python
fantasyTotal = fantasyTotal[(fantasyTotal['Injury'] == 'NA') | (fantasyTotal['Injury'] == 'questionable')]
```

```python
fantasyTotal['CPT FP/$'] = fantasyTotal['DKPt']/fantasyTotal['Salary CPT']
fantasyTotal['Flex FP/$'] = fantasyTotal['DKPt']/fantasyTotal['Salary Flex']
fantasyTotal = fantasyTotal[fantasyTotal['Salary CPT'] != 0]
```
